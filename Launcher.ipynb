# MLP Visualizer

Run the cell below to launch the MLP visualizer right in your browser! Make sure to wait a few seconds after running for the launch button to appear.

```python
import os
import time
import random
import subprocess
import sys
from IPython.display import display, HTML

print("Downloading CPU-optimized packages (this should only take ~20 seconds)...")
# Explicitly force PIP to download the CPU-only version of PyTorch (150MB instead of the massive 2.5GB CUDA GPU version)
subprocess.check_call([sys.executable, "-m", "pip", "install", "torch", "--index-url", "https://download.pytorch.org/whl/cpu", "-q"])
subprocess.check_call([sys.executable, "-m", "pip", "install", "streamlit", "networkx", "plotly", "pandas", "-q"])

port = random.randint(15000, 30000)
print(f"Launching Streamlit explicitly detached on Port {port}...")

log_file = open('streamlit.log', 'w')
process = subprocess.Popen(
    [sys.executable, '-m', 'streamlit', 'run', 'app.py', 
     '--server.port', str(port), 
     '--server.address', '0.0.0.0',
     '--server.enableCORS', 'false',
     '--server.enableXsrfProtection', 'false'],
    stdout=log_file,
    stderr=subprocess.STDOUT,
    start_new_session=True
)

time.sleep(5)

with open('streamlit.log', 'r') as f:
    logs = f.read()
    print("\n--- STREAMLIT RAW STARTUP LOGS ---")
    print(logs)
    print("----------------------------------\n")

url = f"https://datahub.berkeley.edu/user-redirect/proxy/{port}/"
button_html = f"""
<a href="{url}" target="_blank" style="
    display: inline-block; 
    padding: 15px 32px; 
    font-size: 20px; 
    font-family: Arial, sans-serif;
    font-weight: bold; 
    color: white; 
    background-color: #003262;
    border-radius: 8px; 
    text-decoration: none;
    box-shadow: 0 4px 6px rgba(0,0,0,0.1);
">Launch MLP Visualizer</a>
"""
display(HTML(button_html))
```
