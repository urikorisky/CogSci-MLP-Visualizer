# MLP Visualizer

Run the cell below to launch the MLP visualizer right in your browser! Make sure to wait a few seconds after running for the launch button to appear.

```python
import os
import time
import random
import subprocess
import sys
from IPython.display import display, HTML

print("Verifying dependencies (this might take 15 seconds if it's your first time)...")
subprocess.check_call([sys.executable, "-m", "pip", "install", "streamlit", "torch", "networkx", "plotly", "pandas", "-q"])

port = random.randint(15000, 30000)

print(f"Launching Streamlit securely on Port {port}...")
cmd = f"streamlit run app.py --server.port {port} --server.enableCORS false --server.enableXsrfProtection false > streamlit_error.log 2>&1 &"
os.system(cmd)
time.sleep(4)

# Check if the log file contains errors indicating it crashed immediately
with open("streamlit_error.log", "r") as f:
    logs = f.read()
    if "Traceback" in logs or "Error:" in logs or "not found" in logs or "command not found" in logs:
        print("\n--- STREAMLIT CRASHED ---")
        print(logs)
    else:
        url = f"https://datahub.berkeley.edu/user-redirect/proxy/{port}/"
        button_html = f"""
        <a href="{url}" target="_blank" style="
            display: inline-block; 
            padding: 15px 32px; 
            font-size: 20px; 
            font-family: Arial, sans-serif;
            font-weight: bold; 
            color: white; 
            background-color: #003262;
            border-radius: 8px; 
            text-decoration: none;
            box-shadow: 0 4px 6px rgba(0,0,0,0.1);
        ">Launch MLP Visualizer</a>
        """
        display(HTML(button_html))
```
