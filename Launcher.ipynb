# MLP Visualizer

Run the cell below to launch the MLP visualizer right in your browser! Make sure to wait a few seconds after running for the launch button to appear.

```python
import os
import time
import random
import subprocess
import sys
from IPython.display import display, HTML

print("Diagnosing environment...")
env = os.environ.copy()
env["PYTHON_KEYRING_BACKEND"] = "keyring.backends.null.Keyring"

try:
    import torch
    import streamlit
    import networkx
    print("All frameworks located securely.")
except ImportError:
    print("Initial frameworks missing. Installing fresh via Pip...\n")
    subprocess.run([sys.executable, "-m", "pip", "install", "torch", "--index-url", "https://download.pytorch.org/whl/cpu"], env=env)
    subprocess.run([sys.executable, "-m", "pip", "install", "streamlit", "networkx", "plotly", "pandas"], env=env)

port = random.randint(15000, 30000)
print(f"\nLaunching Streamlit explicitly detached on Port {port}...")

# Critically important: stdin=subprocess.DEVNULL violently severs the process from Jupyter's input loop so it doesn't freeze the cell!
log_file = open('streamlit.log', 'w')
process = subprocess.Popen(
    [sys.executable, '-m', 'streamlit', 'run', 'app.py', 
     '--server.port', str(port), 
     '--server.address', '0.0.0.0',
     '--server.enableCORS', 'false',
     '--server.enableXsrfProtection', 'false'],
    stdout=log_file,
    stderr=subprocess.STDOUT,
    stdin=subprocess.DEVNULL,
    start_new_session=True
)

time.sleep(5)

with open('streamlit.log', 'r') as f:
    logs = f.read()
    print("\n--- STREAMLIT RAW STARTUP LOGS ---")
    print(logs)
    print("----------------------------------\n")

url = f"https://datahub.berkeley.edu/user-redirect/proxy/{port}/"
button_html = f"""
<a href="{url}" target="_blank" style="
    display: inline-block; 
    padding: 15px 32px; 
    font-size: 20px; 
    font-family: Arial, sans-serif;
    font-weight: bold; 
    color: white; 
    background-color: #003262;
    border-radius: 8px; 
    text-decoration: none;
    box-shadow: 0 4px 6px rgba(0,0,0,0.1);
">Launch MLP Visualizer</a>
"""
display(HTML(button_html))
```
